```python
import nltk
import numpy as np
import re
from nltk import word_tokenize
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('tagsets')
def Phrases(sentence):
  sentence = word_tokenize(sentence)
  set1=nltk.pos_tag(sentence)
  a=np.empty(shape=len(set1),dtype=str)
  for i in range(len(set1)):
    a[i]=set1[i][1]
  for i in range(len(a)):
    if (i-2)>=0 and a[i]=='N' and a[i-1]=='J' and a[i-2]=='D' :
      s=''
      s=s+set1[i-2][0]+" "+set1[i-1][0]+" "+set1[i][0]
      print(s)
    if (i-1)>=0 and a[i]=='N' and a[i-1]== 'D':
      s=''
      s=s+ set1[i-1][0]+" "+set1[i][0]
      print(s)
    if (i-2)>=0 and a[i]=='N' and a[i-1]== 'D' and a[i-2]=='V' :

      s=''
      s=s+set1[i-2][0]+" "+set1[i-1][0]+" "+set1[i][0]
      print(s)
    if (i+1)<=(len(a)-1)and a[i]=='N' and a[i+1]== 'V' :

      s=''
      s=s+set1[i][0]+" "+set1[i+1][0]
      print(s)
    if (i+2)<=(len(a)-1)  and a[i]=='N' and a[i+1]== 'V' and a[i+2]=='V':

      s=''
      s=s+set1[i][0]+" "+set1[i+1][0]+" "+set1[i+2][0]
      print(s)
    if (i+2)<=(len(a)-1) and (i-1)>=0 and a[i-1]=='N'and a[i]=='V' and a[i+1]== 'R' and a[i+2]=='J':
      s=''
      s=s+set1[i-1][0]+" "+set1[i][0]+" "+set1[i+1][0]+" "+set1[i+2][0]
      print(s)
    if (i+2)<=(len(a)-1) and (i-1)>=0 and a[i-1]=='D'and a[i]=='R' and a[i+1]== 'J' and a[i+2]=='N':
      s=''
      s=s+set1[i-1][0]+" "+set1[i][0]+" "+set1[i+1][0]+" "+set1[i+2][0]
      print(s)
    if (i+2)<=(len(a)-1)  and a[i]=='N' and a[i+1]== 'V' and a[i+2]=='J':

      s=''
      s=s+set1[i][0]+" "+set1[i+1][0]+" "+set1[i+2][0]
      print(s)
    if (i+1)<=(len(a)-1) and a[i]=='N' and a[i+1]== 'N' :

      s=''
      s=s+set1[i][0]+" "+set1[i+1][0]
      print(s)
    if (i+2) <=(len(a)-1) and a[i]=='D'and a[i+1]=='N' and a[i+2]== 'N'  :
      s=''
      s=s+set1[i][0]+" "+set1[i+1][0]+" "+set1[i+2][0]
      print(s)
    if  (i+1)<=(len(a)-1) and a[i]=='J' and a[i+1]=='N' :
      s=''
      s=s+set1[i][0]+" "+set1[i+1][0]
      print(s)
    if  (i+1)<=(len(a)-1) and a[i]=='J' and a[i+1]=='J' :
      s=''
      s=s+set1[i][0]+" "+set1[i+1][0]
      print(s)
    if  (i+1)<=(len(a)-1) and a[i]=='J' and a[i+1]=='V' :
      s=''
      s=s+set1[i][0]+" "+set1[i+1][0]
      print(s)
    if  (i+3)<=(len(a)-1) and a[i]=='J' and a[i+1]=='J' and a[i+2]=='N' and a[i+3]=='N':
      s=''
      s=s+set1[i][0]+" "+set1[i+1][0]+" "+set1[i+2][0]+" "+set1[i+3][0]
      print(s)
    if  (i+3)<=(len(a)-1) and a[i]=='D' and a[i+1]=='J' and a[i+2]=='V' and a[i+3]=='N':
      s=''
      s=s+set1[i][0]+" "+set1[i+1][0]+" "+set1[i+2][0]+" "+set1[i+3][0]
      print(s)
    if  (i+2)<=(len(a)-1) and a[i]=='J' and a[i+1]=='J' and a[i+2]=='N' :
      s=''
      s=s+set1[i][0]+" "+set1[i+1][0]+" "+set1[i+2][0]
      print(s)
    if  (i+2)<=(len(a)-1) and a[i]=='J'  and a[i+1]=='N' and a[i+2]=='N':
      s=''
      s=s+set1[i][0]+" "+set1[i+1][0]+" " +set1[i+2][0]
      print(s)
    if  (i+2)<=(len(a)-1) and a[i]=='D'  and a[i+1]=='J' and a[i+2]=='V':
      s=''
      s=s+set1[i][0]+" "+set1[i+1][0]+" " +set1[i+2][0]
      print(s)
    if  (i+2)<=(len(a)-1) and a[i]=='J'  and a[i+1]=='V' and a[i+2]=='N':
      s=''
      s=s+set1[i][0]+" "+set1[i+1][0]+" " +set1[i+2][0]
      print(s)          
    if (i+2)<=(len(a)-1) and a[i]=='P' and a[i+1]=='M' and a[i+2]=='V':
      s=''
      s=s+set1[i][0]+" "+set1[i+1][0]+" "+set1[i+2][0]
      print(s)
    if (i+1)<=(len(a)-1) and a[i]=='P' and a[i+1]=='N':
      s=''
      s=s+set1[i][0]+" "+set1[i+1][0]
      print(s)
    if (i+1)<=(len(a)-1) and a[i]=='P' and a[i+1]=='V':
      s=''
      s=s+set1[i][0]+" "+set1[i+1][0]
      print(s)
```

```python
print("Input the Text")
sentence=input()
print("""=====
=====""")
print("Possible Phrases")
print(""" 
 
 """)
a_list = nltk.tokenize.sent_tokenize(sentence)
for i in range(len(a_list)):
  sentence=a_list[i]
  sentence = re.sub(r'[?|$|.|!]',r'',sentence)
  sentence = re.sub(r'[^a-zA-Z0-9 ]',r'',sentence)
  Phrases(sentence)




```
